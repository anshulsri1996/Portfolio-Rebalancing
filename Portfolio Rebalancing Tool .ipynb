##We can input the ticker symbol as we wish to change the assets and change the start and end period for which we want to observe the portfolio performance.

#__WITH REBALANCING__

```python
import numpy as np
import pandas as pd
import matplotlib as mlt
import seaborn as sns
from matplotlib import pyplot as plt
import warnings
import yfinance as yf
warnings.filterwarnings("ignore")

initial_capital=1000000
dynamic_capital=0
tickers=['ZGLD.SW','MSFT','REIT','TIP']
initial_allocation={'ZGLD.SW':0.25,'MSFT':0.25,'REIT':0.25,'TIP':0.25} #gold, microsoft, bitcoin, bonds
dynamic_allocation=initial_allocation.copy()
new_allocation={}
start_date='2023-01-01'
end_date='2023-11-30'
ohlc_data={}
amount_dict={}
for ticker in tickers:
    ohlc_data[ticker]=yf.download(ticker, start=start_date, end=end_date)
    ohlc_data[ticker].reset_index(inplace=True)
    ohlc_data[ticker].rename(columns={'index': 'Date'}, inplace=True)
    units=initial_allocation[ticker]*initial_capital/int(ohlc_data[ticker]['Adj Close'].iloc[0])
    amount_dict[ticker]=['00',initial_allocation[ticker]*initial_capital,units]
    print(f'The initial capital invested in {ticker} is {amount_dict[ticker][1]} USD and the units purchased at t=0 is {amount_dict[ticker][2]}')
    print(f'\n{amount_dict}')


#rebalancing @ month start
def print_month_end_allocation(amount_dict,c):
    amount_temp={}
    new_temp={}
    dynamic_capital=0
    for ticker in tickers:
        for i in range(c,len(ohlc_data[ticker])):
            if ohlc_data[ticker]['Date'][i-1].date().strftime("%m")!=ohlc_data[ticker]['Date'][i].date().strftime("%m"):
                #amount_temp[ticker] = [month, amount in portfolio, units]
                amount_temp[ticker] =[ohlc_data[ticker]['Date'][i].date().strftime("%m"),\
                                     int(amount_dict[ticker][2]*ohlc_data[ticker]['Adj Close'].loc[i]),\
                                     amount_dict[ticker][2]]
                date=str(ohlc_data[ticker]['Date'][i].date().strftime("%Y-%m-%d"))
                #print(f'The capital invested in {ticker} is {amount_temp[ticker][1]} USD and the units held on {date} is {amount_temp[ticker][2]}')
                break
        dynamic_capital+=amount_temp[ticker][1]
        #print(date, i, ticker, dynamic_capital)
    
    #dynamic_allocation[ticker]=[% allocation, units, close price, value of ticker in portfolio]
    for ticker in tickers:
        dynamic_allocation[ticker]=[amount_temp[ticker][1]/dynamic_capital,\
                                    amount_dict[ticker][2],\
                                    int(ohlc_data[ticker]['Adj Close'].loc[i]),\
                                    amount_temp[ticker][1]]
        new_allocation[ticker]=[0.25,\
                                dynamic_capital*0.25/int(ohlc_data[ticker]['Adj Close'].loc[i]),\
                                int(ohlc_data[ticker]['Adj Close'].loc[i]),\
                                dynamic_capital*0.25]
        new_temp[ticker] =[ohlc_data[ticker]['Date'][i].date().strftime("%m"),\
                           dynamic_capital*0.25,\
                           dynamic_capital*0.25/int(ohlc_data[ticker]['Adj Close'].loc[i])]
        
    
    print(f'\nThe dynamic allocation after date {date}  is {dynamic_allocation}')
    print(f'\nThe new allocation after re-balancing on date {date}  is {new_allocation}')
    print(f'\nThe total capital after date {date}  is {dynamic_capital}')
    c=i+3
    
    return new_temp,c

c=1
while c<210:
    p=print_month_end_allocation(amount_dict,c)
    amount_dict=p[0]
    c=p[1]
```
